```python
import os
import numpy as np 
from   numpy import random
import matplotlib.pyplot as plt
from   PIL   import Image, ImageOps
from tensorflow.keras.applications import VGG16
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.vgg16 import decode_predictions, preprocess_input
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.activations import relu, softmax
from tensorflow.keras.optimizers import SGD, Adam, RMSprop
from tensorflow.keras.losses import categorical_crossentropy
import tensorflow as tf
```

# Задание 1

## Обучение модели

Загрузим модель VGG16

```python
shivams_model = VGG16(include_top = True)
```

```python
shivams_model.summary()
```

Создаем генератор данных, который будет проводить аугментацию данных

```python
idg = ImageDataGenerator(preprocessing_function = preprocess_input,horizontal_flip = True,validation_split=0.2)
```

 Генератор сам распознает имеющиеся классы

```python
train_generator = idg.flow_from_directory("../data/dataset/training_set",batch_size=(128),target_size=(224,224),subset="training")
val_generator = idg.flow_from_directory("../data/dataset/training_set",batch_size=(128),target_size=(224,224),subset="validation")
```

```python
x = shivams_model.layers[-3].output
x = BatchNormalization(trainable = False)(x)
x = Dense(2, activation = softmax)(x) 
```

```python
final_model = Model(inputs = shivams_model.input, outputs = x)
```

```python
final_model.summary()
```

```python
final_model.compile(RMSprop(), categorical_crossentropy,['acc'])
```

Обучаем модель с использованием генераторов

```python
model_hist = final_model.fit(train_generator, validation_data = val_generator, epochs=10)
```

Функция создания тестовых примеров

```python
def create_test_samples(img_path, image_list ):
  test_data = []
  for img in image_list:
    im1 = cv2.imread(os.path.join(img_path,img))
    im1 = cv2.resize(im1,(224,224))
    test_data.append(im1)
  test_data = np.array(test_data)
  return test_data
```

Примеры котов и собак для тестирования

```python
test_set1 = create_test_samples("dataset/training_set/cats",test_dir_cat[100:400])
test_set2 = create_test_samples("dataset/training_set/dogs",test_dir_dog[:300])
test_set1.shape,test_set2.shape
```

Функция для предсказывания класса изображения

```python
def predict_cat_or_dog(test_img):
  pred = final_model.predict(test_img)
  pred = pred.argmax(axis=1)
  return pred
```

Результат 300 предсказаний для котов (0 - кот, 1 - собака)

```python
cat_test_result = predict_cat_or_dog(test_set1)
print(cat_test_result)
```

Результат 300 предсказаний для собак (0 - кот, 1 - собака)

```python
dog_test_res = predict_cat_or_dog(test_set2)
print(dog_test_res)
```

Точность для собак

```python
predicted = np.count_nonzero(dog_test_res)
accuracy = (predicted / 300)*100
accuracy
```

Точность для кошек

```python
predicted = np.count_nonzero(cat_test_result)
accuracy = ((300 - predicted) / 300)*100       
accuracy
```
